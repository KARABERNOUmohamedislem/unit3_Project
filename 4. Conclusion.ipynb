## 2. Analysis: 

```python
# To-Do create a summary about each cluster : 
# percentages of each target (actual Category) within each cluster 
# + Percentages of each predicted class within the cluster 
# results should be something like : 
# For clusterk : {'true_class0_percetage': 0.0,
#                 'true_class1_percetage': 0.0,
#                  'true_class2_percetage': 0.0,
#                   'true_class3_percetage': 0.0,
#                   'true_class4_percetage': 0.0,
#                   'predicted_class0_percetage': 0.0,
#                   'predicted_class1_percetage': 0.0,
#                   'predicted_class2_percetage': 0.0,
#                   'predicted_class3_percetage': 0.0,
#                    'predicted_class4_percetage': 0.0}
clusters = []
for i in range(5):
    clusters.append({})
    cluster_count = len(df[df.cluster == i])
    if cluster_count == 0:
        clusters[i]["true_class0_percetage"] = 0
        clusters[i]["true_class1_percetage"] = 0
        clusters[i]["true_class2_percetage"] = 0
        clusters[i]["true_class3_percetage"] = 0
        clusters[i]["true_class4_percetage"] = 0
        clusters[i]["predicted_class0_percetage"] = 0
        clusters[i]["predicted_class1_percetage"] = 0
        clusters[i]["predicted_class2_percetage"] = 0
        clusters[i]["predicted_class3_percetage"] = 0
        clusters[i]["predicted_class4_percetage"] = 0
    else:
        clusters[i]["true_class0_percetage"] = len(df[(df.cluster == i) & (df.Category == 0)]) / cluster_count
        clusters[i]["true_class1_percetage"] = len(df[(df.cluster == i) & (df.Category == 1)]) / cluster_count
        clusters[i]["true_class2_percetage"] = len(df[(df.cluster == i) & (df.Category == 2)]) / cluster_count
        clusters[i]["true_class3_percetage"] = len(df[(df.cluster == i) & (df.Category == 3)]) / cluster_count
        clusters[i]["true_class4_percetage"] = len(df[(df.cluster == i) & (df.Category == 4)]) / cluster_count
        clusters[i]["predicted_class0_percetage"] = len(df[(df.cluster == i) & (df.Predicted_class == 0)]) / cluster_count
        clusters[i]["predicted_class1_percetage"] = len(df[(df.cluster == i) & (df.Predicted_class == 1)]) / cluster_count
        clusters[i]["predicted_class2_percetage"] = len(df[(df.cluster == i) & (df.Predicted_class == 2)]) / cluster_count
        clusters[i]["predicted_class3_percetage"] = len(df[(df.cluster == i) & (df.Predicted_class == 3)]) / cluster_count
        clusters[i]["predicted_class4_percetage"] = len(df[(df.cluster == i) & (df.Predicted_class == 4)]) / cluster_count
        
clusters
```
